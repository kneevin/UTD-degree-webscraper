```python
import pandas as pd
import numpy as np
import re
```

```python
catalog_file = 'data/DATA LABELED LINK.csv'
all_courses_file = 'course codes.csv'
```

```python
catalog = pd.read_csv(catalog_file)
all_courses = pd.read_csv(all_courses_file)
```

```python
catalog['Updated File Name'] = 'updated ' + catalog['Course File Name']
main_table = pd.DataFrame(catalog[['Title','Updated File Name']])
```

```python
cols = all_courses.to_numpy().flatten()
# pd.concat([main_table, pd.Series(cols)])
cols
```

```python
main_table = catalog.reindex(columns=['Title', 'Updated File Name'] + ['Count'] + cols.tolist())
main_table = main_table.fillna(0)
main_table
```

```python
for i, file_name in enumerate(main_table['Updated File Name']):
    count = 0
    try:
        for c in pd.read_csv(file_name)['Course Code']:
            main_table.loc[i, c] = 1
            count += 1
    except:
        print(file_name + ' did not read any inputs.')
    main_table.loc[i, 'Count'] = count
```
