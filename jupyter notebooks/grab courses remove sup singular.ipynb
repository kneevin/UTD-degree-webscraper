```python
# passed a URL, returns a list of all courses
def findCourses(URL):
  temp = []
  page = requests.get(URL)
  soup = BeautifulSoup(page.text,'html.parser')
  classes = soup.find_all(class_='coursebook-info') # classes all have the class name of 'coursebook-info'
  for c in classes:
    for s in list(c.next_siblings)[1:]: # the first element after the course number is always the title, hence remove anything after that
      try:
        s.decompose()
      except:
        continue
  for c in classes:
    try:
      temp.append(c.parent.text)
    except: 
      continue
  return temp
```

```python
findCourses('https://catalog.utdallas.edu/2022/undergraduate/programs/ah/latin-american-studies')
```
