```python
import numpy as np
import pandas as pd
import time
import requests
import re
from bs4 import BeautifulSoup
from google.colab import files
```

```python
# literal gold right here
course_regex = re.compile('^(or )?([A-Z]+) ([0-9|A-Z]+) (.+)$')
```

```python
# create soup, given a URL
def create_soup(URL_):
  page = requests.get(URL_)
  soup = BeautifulSoup(page.text, 'html.parser')
  return soup.find_all('p')
```

```python
# getting rid of footnote links at the end of certain courses
def remove_footnotes(class_soup):
  for c in class_soup:
    for child in c.children:
      try:
        if 'footnote' in child['href']:
          child.decompose()
      except:
        continue
```

```python
# using regex to grab all courses, returns list of courses
def find_courses(class_soup, reg):
  temp = []
  for c in class_soup:
      if reg.search(c.text):
        temp.append(c.text)
  return temp
```

```python
df = pd.read_csv('DATA NO COURSES.csv')
df['Course File Name'] = [np.NaN for i in range(len(df['URL Links']))]
```

```python
df.head()
```

```python
for i, URL in enumerate(df['Catalog Page']):
  try:
    classes = create_soup(URL)
    remove_footnotes(classes)
    df.loc[i, 'Course File Name'] = f"courses/{df.loc[i,'Title']}.csv"
    pd.Series(find_courses(classes,course_regex)).to_csv(df.loc[i, 'Course File Name'], index=False, header=False)
    print(df.loc[i])
  except:
    print('NOTHING WAS ADDED FOR', df.loc[i, 'Title'])
  time.sleep(30)
  df.to_csv('output.csv', index=False)
```
